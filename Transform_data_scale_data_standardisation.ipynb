```python
import pandas as pd
import scipy

import matplotlib.pyplot as plt
import seaborn as sb
from matplotlib import rcParams

import sklearn 
from sklearn import preprocessing
from sklearn.preprocessing import scale
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4
sb.set_style('whitegrid')
```

```python
file_name = 'mtcars.csv'
cars = pd.read_csv(file_name)

cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
```

```python
mpg = cars.mpg
plt.plot(mpg)
```

```python
mpg.describe()
```

```python
# Scaling the mpg value to 0,1, Minmaxscalaer, if no args passed, defaults to 0.1
mpg_matrix = mpg.reshape(-1,1)
scaled = preprocessing.MinMaxScaler()
scaled_mpg = scaled.fit_transform(mpg_matrix)
plt.plot(scaled_mpg)
```

```python
# To rescale to given range , pass feature_range in MinMaxscalaer
mpg_matrix = mpg.reshape(-1,1)
scaled = preprocessing.MinMaxScaler(feature_range=(0,10))
scaled_mpg = scaled.fit_transform(mpg_matrix)
plt.plot(scaled_mpg)
```

```python
# Belwo is standarisation, if with_mean-False and with_std is false , the same values as mpg will be used
std_mpg = scale(mpg,axis=0,with_mean=False,with_std=False)
plt.plot(std_mpg)
```

```python
# But if those values are not provided, then by default ,standarisation happens with mean set to 0
std_mpg = scale(mpg)
plt.plot(std_mpg)
```
