```python
# Import libaries
import pandas as pd
import numpy as np

from pandas import Series,DataFrame
```

```python
# Adding value of nan to a variable
missing = np.nan
```

```python
# Creating a series with few nan values
series_obj = Series(['row 1','row 2',missing,'row 4','row 5','row 6',missing,'row 8'])
print(series_obj)
```

```python
# To know which index are nan
series_obj.isnull()
```

```python
# Creating a randon DF
np.random.seed(25)
df_obj = DataFrame(np.random.randn(36).reshape(6,6))
print(df_obj)
#df_obj1 = DataFrame(np.random.rand(36).reshape(6,6))
#df_obj1
```

```python
# Manually adding few nan values
df_obj.iloc[3:5,0] = missing
df_obj.iloc[1:4,5] = missing

df_obj
```

```python
# Repalce the nan values with 0
filled_df = df_obj.fillna(0)
filled_df
```

```python
# Fill the missing values with a dictinary of vales
# HEre the column 0's nan values are replaced with 1 , column 5 nan val replaced with 2.5
filled_df = df_obj.fillna({0:1.0,5:2.5})
filled_df
```

```python
# Fill the missing values with a dictinary of vales
# The raplce happens only for columns using dic, row repalce doesnot work
filled_df = df_obj.fillna({0:1.0,5:2.5},axis=1)
filled_df
```

```python
# Ffill fils missing value with last value in the column
filled_df = df_obj.ffill()
filled_df
```

```python
# To find the sum of mising values
# The values are w.r.t cols, col 0 has 3 nan values and col 5 has 3 missing vals
df_obj.isnull().sum()
```

```python
# to drop all rows which have missing values, dropna
df_no_naobj = df_obj.dropna()
df_no_naobj
```

```python
# To drop all columns with missing values , axis = 1
df_no_naobj = df_obj.dropna(axis=1)
df_no_naobj
```

```python
# to drop only if all rows are nan
df_allnan = df_obj.dropna(how='all')
df_allnan
```
