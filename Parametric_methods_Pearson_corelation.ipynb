```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb
import scipy 
from scipy.stats.stats import pearsonr
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8,4
sb.set_style('whitegrid')
```

```python
file_name = 'mtcars.csv'
cars = pd.read_csv(file_name)

cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
```

```python
sb.pairplot(cars)
```

```python
X =cars[['mpg','hp','qsec','wt']]
sb.pairplot(X)
```

```python
# Creating individual datapoints
mpg = cars['mpg']
hp = cars['hp']
qsec = cars['qsec']
wt = cars['wt']
```

```python
# Finding out pearson coeficient 
pearsonr_coeficient, p_value = pearsonr(mpg,hp)
print("Pearson COefiicient value %0.3f" %(pearsonr_coeficient))
```

```python
pearsonr_coeficient, p_value = pearsonr(mpg,qsec)
print("Pearson COefiicient value %0.3f" %(pearsonr_coeficient))
```

```python
pearsonr_coeficient, p_value = pearsonr(mpg,wt)
print("Pearson COefiicient value %0.3f" %(pearsonr_coeficient))
```
