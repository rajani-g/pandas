```python
import numpy as np
import pandas as pd
from numpy.random import  randn

import matplotlib.pyplot as plt
from matplotlib import rcParams

import seaborn as sb
```

```python
# The below line is to indictae notebook to print here 
%matplotlib inline

rcParams['figure.figsize'] = 5,4
sb.set_style('whitegrid')
```

```python
X = range(1,10)
Y = [1,2,3,4,0,4,3,2,1]
```

```python
plt.plot(X,Y)
plt.show()
```

```python
file_name = 'mtcars.csv'
cars = pd.read_csv(file_name)

cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
mpg = cars['mpg']

mpg.head()
```

```python
# Line charts
mpg.plot()
```

```python
# Line charts
df = cars[['cyl','wt','mpg']]

df.plot()
```

```python
# Bar chart
plt.bar(X,Y)
```

```python
# Using pandas to plot
mpg.plot(kind='bar')
```

```python
mpg.plot(kind='barh')
```

```python
# Pie chart
X=[1,2,3,4,0.5]
plt.pie(X)
plt.show()
```

```python
# Save the fig
plt.savefig('piechart.jpeg')
plt.show()

```
