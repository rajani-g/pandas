```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import sklearn
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import scale

import sklearn.metrics as sm
from sklearn import datasets
from sklearn.metrics import  confusion_matrix,classification_report
```

```python
%matplotlib inline
plt.figure(figsize=(7,4))
```

```python
iris = datasets.load_iris()

X=scale(iris.data)
y = pd.DataFrame(iris.target)

variable_name = iris.feature_names
X[0:10,]
```

```python
clustering = KMeans(n_clusters=3,random_state=5)
clustering.fit(X)
```

```python
iris_df = pd.DataFrame(iris.data)
iris_df.columns = ['sepal_length', 'sepal_width', 'petal_length','petal_width']
print(iris_df.columns)

y.columns = ['target']
```

```python
color_theme = np.array(['darkgray','lightsalmon','powderblue'])
```

```python
# Actual vales from ds
plt.subplot(1,2,1)
plt.scatter(x=iris_df.petal_length,y=iris_df.petal_width,c=color_theme[iris.target],s=50)
plt.title('Ground truth classification')

# Predicted labels using K-means 
plt.subplot(1,2,2)
plt.scatter(x=iris_df.petal_length,y=iris_df.petal_width,c=color_theme[clustering.labels_],s=50)
plt.title('K means classification')

# If you see the colors are npt matching, that means the label names are wrongly asigned, can be fixed with choose
```

```python
# Reassiging label names
relabel = np.choose(clustering.labels_,[2,0,1]).astype(np.int64)

plt.subplot(1,2,1)
plt.scatter(x=iris_df.petal_length,y=iris_df.petal_width,c=color_theme[iris.target],s=50)
plt.title('Ground truth classification')

plt.subplot(1,2,2)
plt.scatter(x=iris_df.petal_length,y=iris_df.petal_width,c=color_theme[relabel],s=50)
plt.title('K means classification')


```

```python
# checking how accurately it predicted , so good model = high precision + hisgh recall
print(classification_report(y,relabel))
```
