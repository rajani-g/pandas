```python
import numpy as np
import pandas as pd 

import matplotlib.pyplot as plt
import seaborn as sb
from matplotlib.pyplot import rcParams

import scipy
from scipy.stats.stats import spearmanr
# lib for chi-square test
from scipy.stats import chi2_contingency
```

```python
# Spearsman works on ordinal pairs,which can be categorised
#--- non linear
# -- non -normally distributed
# p < 0.05 -- Reject Null hypothesis -- and it means the 2 variables are corelated
# p > 0.05 -- Accept null hypothesis -- variables are independent
```

```python
%matplotlib inline

file_name = 'mtcars.csv'
cars = pd.read_csv(file_name)

cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']

```

```python
sb.pairplot(cars)
```

```python
X=cars[['cyl','vs','am','gear']]
sb.pairplot(X)
```

```python
cyl = cars['cyl']
vs = cars['vs']
am = cars['am']
gear = cars['gear']

spearman_coeficient,p_value = spearmanr(cyl,vs)
print("Pearson coeficient %0.3f" %spearman_coeficient)
```

```python
spearman_coeficient,p_value = spearmanr(cyl,am)
print("Pearson coeficient %0.3f" %spearman_coeficient)
```

```python
spearman_coeficient,p_value = spearmanr(cyl,gear)
print("Pearson coeficient %0.3f" %spearman_coeficient)
```

```python
# fr creating chi-square , we need to have cross tab
table = pd.crosstab(cyl,am)
print(table)

chi2_contingency(table.values)
chi2,p,dof,expected = chi2_contingency(table.values)
print("Chi-square statistic %0.3f p-value %0.3f" %(chi2,p))
```

```python
table = pd.crosstab(cyl,vs)
print(table)

chi2,p,dof,expected = chi2_contingency(table.values)
print("Chi-square statistic %0.3f p-value %0.3f" %(chi2,p))
```

```python

table = pd.crosstab(cyl,gear)
print(table)

chi2,p,dof,expected = chi2_contingency(table.values)
print("Chi-square statistic %0.3f p-value %0.3f" %(chi2,p))
```
