```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb

import networkx as nx
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 7,4
sb.set_style('whitegrid')
```

```python
# Creating a blank graph
G = nx.Graph()
nx.draw(G)
```

```python
# Adding 1 node
G.add_node(1)
nx.draw(G)
```

```python
# Adding multiple nodes
G.add_nodes_from([2,3,4,5,6,8,9,12,15,16,200])
nx.draw(G)
```

```python
# adding edges between nodes
G.add_edges_from([(2,4),(2,6),(2,8)])
nx.draw(G)
```

```python
# drawing in circular chape
nx.draw_circular(G)
```

```python
nx.draw_spring(G)
```

```python
# adding labels
nx.draw_circular(G,node_color='bisque',with_labels=True)
```

```python
# removing a node
G.remove_node(1)
nx.draw_circular(G,node_color='bisque',with_labels=True)
```

```python
# printing the summary of G, number of nodes,edges count etc
summary_stats = nx.info(G)
print(summary_stats)
```

```python
# print the degree of each node, that is show number of nodes each nde is conencted to  
print(nx.degree(G))
```

```python
# till now we added graphs manually, the beloe are few graph generators
G=nx.complete_graph(25)
nx.draw(G,node_color='bisque',with_labels=True)
```

```python
# gnc graph
G=nx.gnc_graph(7,seed=25)
nx.draw(G,node_color='bisque',with_labels=True)
```

```python
# ego graph
ego_G = nx.ego_graph(G,3,radius=5)
nx.draw(G,node_color='bisque',with_labels=True)
```
