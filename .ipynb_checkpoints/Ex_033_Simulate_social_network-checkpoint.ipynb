```python
import pandas as pd
import numpy as np

import networkx as nx

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4
sb.set_style('whitegrid')
```

```python
# cerate a direct graph
DG = nx.gn_graph(7,seed=25)
for line in nx.generate_edgelist(DG,data=False):print line
    
```

```python
print DG.node[0]
```

```python
# adding nodes
DG.node[0]['name'] = 'Alice'
print DG.node[0]
```

```python
DG.node[1]['name'] = 'Bob'
DG.node[2]['name'] = 'Maria'
DG.node[3]['name'] = 'rajani'
DG.node[4]['name'] = 'Sai'
DG.node[5]['name'] = 'Pranawi'
DG.node[6]['name'] = 'Anjali'

```

```python
DG.add_nodes_from([(0,{'age':25}),(1,{'age':25}),(2,{'age':25}),(3,{'age':25}),(4,{'age':25}),(5,{'age':25}),(6,{'age':25})])
print DG.node[0]
```

```python
nx.draw_circular(DG,node_color='bisque',with_labels=True)
```

```python
# adding names as labels
labeldictionary = {0:'Alice',1:'Bob'}
nx.draw_circular(DG,node_color='bisque',with_labels=True,labels=labeldictionary)
```

```python
# creating un directed grpah
G=DG.to_undirected()
nx.draw_spectral(G,labels=labeldictionary,with_labels=True)
```
